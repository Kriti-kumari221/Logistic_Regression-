```python
data=data.drop(["Unnamed: 32","id"],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
data["diagnosis"]=le.fit_transform(data["diagnosis"])
data["diagnosis"]
```

```python
data.head()
```

```python
X=data.drop("diagnosis",axis=1)
```

```python
y=data["diagnosis"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
```

```python
pipeline=Pipeline([
    ('scaler',StandardScaler()),
    ('nb',GaussianNB())
])
pipeline.fit(x_train,y_train)
```

```python
y_train_pred=pipeline.predict(x_train)
y_test_pred=pipeline.predict(x_test)
```

```python
print(y_train_pred)
print(y_test_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
training_score=accuracy_score(y_train,y_train_pred)
testing_score=accuracy_score(y_test,y_test_pred)
y_pred_proba_nb=pipeline.predict_proba(x_test)[:, 1]
```

```python
print("\nNaÃ¯ve Bayes:")
print("Training Score: ",training_score)
print("Testing Score: ",testing_score)
```

```python
results_df = pd.DataFrame({
    'Actual Label': y_test,
    'Predicted Label': y_test_pred,
    'Predicted Probability': y_pred_proba_nb
})

results_df.head()
```

```python
from sklearn.metrics import confusion_matrix, classification_report,roc_auc_score,roc_curve
```

```python
con_mat=confusion_matrix(y_test,y_test_pred)
print(con_mat)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
# Heatmap visualization for confusion matrix
display=ConfusionMatrixDisplay(con_mat)
display.plot()
```

```python
print("\t\tNaÃ¯ve Bayes Classification Report:")
print(classification_report(y_test,y_test_pred))
```

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(y_test, y_test_pred)
# Compute the Area Under the Curve (AUC) for the ROC curve
roc_auc = auc(fpr, tpr)
# Compute Youden's J statistic for each threshold
youden_j = tpr - fpr
optimal_threshold_index = np.argmax(youden_j)
optimal_threshold = thresholds[optimal_threshold_index]

print(f"Optimal Threshold: {optimal_threshold:.4f}")

# Plot the ROC curve with the optimal threshold marked
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.scatter(fpr[optimal_threshold_index], tpr[optimal_threshold_index], color='red', marker='o', label=f'Optimal Threshold = {optimal_threshold:.4f}')
plt.plot([0, 1], [0, 1], color='grey', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
